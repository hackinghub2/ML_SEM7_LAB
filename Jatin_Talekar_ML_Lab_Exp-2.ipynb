```python
Sex=pd.get_dummies(df["Sex"])
Sex.head()
```

```python
Sex=pd.get_dummies(df["Sex"])
Sex.head()
```

```python
embark=pd.get_dummies(df["Embarked"])
embark.head()
```

```python
embark=pd.get_dummies(df["Embarked"],drop_first=True)
embark.head()
```

```python
pcl=pd.get_dummies(df["Pclass"],drop_first=True)
pcl.head()
```

```python
df=pd.concat([df,Sex,embark,pcl],axis=1)
df.head()
```

```python
df.drop(["Sex","Embarked","PassengerId","Name","Ticket","Fare"],axis=1,inplace=True)
df.head()
```

```python
df.dtypes
```

```python
df.columns = df.columns.astype(str)
```

```python
X=df.drop("Survived",axis=1)
y=df["Survived"]
```

```python
X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.20, random_state=42)
```

```python
model=LogisticRegression()
model.fit(X_train,y_train)
```

```python
predictions=model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,predictions)
```

```python
from sklearn import metrics
print(metrics.classification_report(y_test, predictions))
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test,predictions)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from sklearn import metrics
# Assuming y_test and predictions are your actual and predicted values
cm = metrics.confusion_matrix(y_test, predictions, labels=[1, 0])
df_cm = pd.DataFrame(cm, index=[i for i in ["1", "0"]],
 columns=[i for i in ["Predict 1", "Predict 0"]])
plt.figure(figsize=(7, 5))
sns.heatmap(df_cm, annot=True, fmt='g', cmap='viridis') # Change 'viridis' to any color map you like
# Show the plot
plt.show()
```

```python
coefficients = model.coef_[0]
feature_importance = pd.DataFrame({
 'Feature': X.columns,
 'Coefficient': coefficients
})
feature_importance['AbsCoefficient'] = feature_importance['Coefficient'].abs()
feature_importance = feature_importance.sort_values(by='AbsCoefficient', ascending=False)
print(feature_importance[['Feature', 'Coefficient']])
plt.figure(figsize=(10, 6))
plt.barh(feature_importance['Feature'], feature_importance['Coefficient'], color='red')
plt.xlabel('Coefficient Value')
plt.title('Feature Importance from Logistic Regression')
plt.gca().invert_yaxis()
plt.show()
```
